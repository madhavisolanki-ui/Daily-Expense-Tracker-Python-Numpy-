# **Project Option 3: Daily Expense Tracker**

6.1 Problem Statement Create a program to track daily expenses and analyze spending.

```python
import numpy as np

# Input number of days
days = int(input("Enter number of days: "))

expenses_list = []

# Take daily expenses as input
for i in range(days):
    expense = float(input("Enter expense: "))
    expenses_list.append(expense)

# Convert list to NumPy array
expenses_array = np.array(expenses_list)

# Calculations
total_expense = expenses_array.sum()
average_expense = expenses_array.mean()
highest_expense = expenses_array.max()

# Output
print("\n--- Daily Expense Summary ---")
print("Expenses:", expenses_array)
print("Total Expense:", total_expense)
print("Average Expense:", average_expense)
print("Highest Expense:", highest_expense)

# Optional warning using condition
if average_expense > 1000:
    print("Warning: Average expense is high!")

```
